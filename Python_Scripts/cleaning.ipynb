```python
import pandas as pd
```

```python
#read in january data
jan1 = pd.read_csv('../Resources/202401-citibike-tripdata_1.csv')
jan2 = pd.read_csv('../Resources/202401-citibike-tripdata_2.csv')
```

```python
#read in february data
feb1 = pd.read_csv('../Resources/202402-citibike-tripdata_1.csv')
feb2 = pd.read_csv('../Resources/202402-citibike-tripdata_2.csv')
feb3 = pd.read_csv('../Resources/202402-citibike-tripdata_3.csv')
```

```python
#read in march data
mar1 = pd.read_csv('../Resources/202403-citibike-tripdata_1.csv')
mar2 = pd.read_csv('../Resources/202403-citibike-tripdata_2.csv')
mar3 = pd.read_csv('../Resources/202403-citibike-tripdata_3.csv')
```

```python
#consolidate jan data
jan = pd.concat([jan1, jan2], axis = 0)
```

```python
#consolidate feb data
feb  = pd.concat([feb1, feb2, feb3], axis = 0)
```

```python
#consolidate mar data
mar = pd.concat([mar1, mar2, mar3], axis = 0)
```

```python
#add month number to each dataframe
jan['month'] = 1
feb['month'] = 2
mar['month'] = 3
```

```python
#find the most used  lat and long for each start station name
grouped_lat = jan[['start_station_name', 'start_lat']].groupby('start_station_name')
grouped_lng = jan[['start_station_name', 'start_lng']].groupby('start_station_name')
pop_lat = grouped_lat['start_lat'].agg(lambda x: x.mode().iloc[0])
pop_lng = grouped_lng['start_lng'].agg(lambda x: x.mode().iloc[0])
```

```python
#map the lat and long to each row
jan['pop_lat'] = jan['start_station_name'].map(pop_lat)
jan['pop_lng'] = jan['start_station_name'].map(pop_lng)
```

```python
#find the most used  lat and long for each start station name
#map the lat and long to each row
grouped_lat = feb[['start_station_name', 'start_lat']].groupby('start_station_name')
grouped_lng = feb[['start_station_name', 'start_lng']].groupby('start_station_name')
pop_lat = grouped_lat['start_lat'].agg(lambda x: x.mode().iloc[0])
pop_lng = grouped_lng['start_lng'].agg(lambda x: x.mode().iloc[0])
feb['pop_lat'] = feb['start_station_name'].map(pop_lat)
feb['pop_lng'] = feb['start_station_name'].map(pop_lng)
```

```python
#find the most used  lat and long for each start station name
#map the lat and long to each row
grouped_lat = mar[['start_station_name', 'start_lat']].groupby('start_station_name')
grouped_lng = mar[['start_station_name', 'start_lng']].groupby('start_station_name')
pop_lat = grouped_lat['start_lat'].agg(lambda x: x.mode().iloc[0])
pop_lng = grouped_lng['start_lng'].agg(lambda x: x.mode().iloc[0])
mar['pop_lat'] = mar['start_station_name'].map(pop_lat)
mar['pop_lng'] = mar['start_station_name'].map(pop_lng)
```

```python
#output each csv and a concatted one
all_2024 = pd.concat([jan, feb, mar])
all_2024.to_csv('all_2024.csv', index = False)
# jan.to_csv('jan_final.csv', index = False)
# feb.to_csv('feb_final.csv', index = False)
# mar.to_csv('mar_final.csv', index = False)
```
